```python
print("00_count_nums.md")
def get_user_numbers():
    """
    Gets numbers from user until blank input
    Returns: List of integers entered by user
    """
    numbers = []
    while True:
        user_input = input("Enter a number (press Enter to finish): ")
        if user_input == "":
            break
        try:
            numbers.append(int(user_input))
        except ValueError:
            print("Please enter a valid number!")
    return numbers

def count_numbers(numbers):
    """
    Counts frequency of each number in list
    Returns: Dictionary with numbers as keys and counts as values
    """
    frequency = {}
    for num in numbers:
        if num in frequency:
            frequency[num] += 1
        else:
            frequency[num] = 1
    return frequency

def print_frequencies(frequency_dict):
    """
    Prints number frequencies in sorted order
    """
    print("\nNumber frequencies:")
    for num in sorted(frequency_dict.keys()):
        print(f"{num} appears {frequency_dict[num]} time(s)")

def main():
    """
    Main program flow
    """
    print("Number Frequency Counter")
    print("Enter numbers one by one, press Enter when done\n")

    numbers = get_user_numbers()
    if not numbers:
        print("No numbers entered!")
        return

    freq = count_numbers(numbers)
    print_frequencies(freq)

if __name__ == "__main__":
    main()
```

```python
print("01_phonebook")
def create_phonebook():
    """
    Creates a phonebook by taking user input
    Returns: Dictionary of name-number pairs
    """
    phonebook = {}
    print("\nAdd contacts to phonebook (leave name blank to finish)")

    while True:
        name = input("Enter contact name: ").strip()
        if not name:  # Exit if name is empty
            break

        # Validate phone number input
        while True:
            number = input(f"Enter {name}'s phone number: ").strip()
            if number:  # Only accept non-empty numbers
                phonebook[name] = number
                print(f"âœ… {name} added to phonebook\n")
                break
            print("âŒ Phone number cannot be empty!")

    return phonebook

def display_phonebook(phonebook):
    """
    Displays all contacts in alphabetical order
    """
    if not phonebook:
        print("\nPhonebook is empty!")
        return

    print("\nðŸ“ž Phonebook Contacts:")
    for name in sorted(phonebook.keys()):
        print(f"{name:20} â†’ {phonebook[name]}")

def lookup_contact(phonebook):
    """
    Allows searching contacts by name
    """
    print("\nðŸ” Contact Lookup")
    while True:
        name = input("Enter name to search (blank to exit): ").strip()
        if not name:
            break

        if name in phonebook:
            print(f"Contact found: {name} - {phonebook[name]}")
        else:
            print(f"âŒ {name} not found in phonebook")

        print()  # Add empty line for spacing

def main():
    """
    Main program function
    """
    print("ðŸŒŸ Phonebook Application ðŸŒŸ")
    phonebook = create_phonebook()
    display_phonebook(phonebook)
    lookup_contact(phonebook)
    print("\nThank you for using Phonebook App!")

if __name__ == '__main__':
    main()
```

```python
print("2")
def main():
    # Fruit prices dictionary
    fruits = {
        'apple': 1.5,
        'durian': 50,
        'jackfruit': 80,
        'kiwi': 1,
        'rambutan': 1.5,
        'mango': 5
    }

    total = 0.0

    print("ðŸ›’ Welcome to the Fruit Shop Calculator!")
    print("Enter quantity for each fruit (press Enter for 0)\n")

    for fruit, price in fruits.items():
        while True:
            try:
                qty_input = input(f"How many {fruit} @ ${price} each? ")
                quantity = int(qty_input) if qty_input else 0
                if quantity >= 0:
                    total += price * quantity
                    break
                print("Please enter positive number")
            except ValueError:
                print("Invalid input! Enter number or press Enter")

    print(f"\nðŸ’° Your total bill is: ${total:.2f}")

if __name__ == '__main__':
    main()
```

```python
print("03_powerful_passwords")
from hashlib import sha256

def login(email, stored_logins, password_to_check):
    """
    Authenticates user by comparing password hash with stored hash

    Args:
        email: User's email address
        stored_logins: Dictionary of {email: hashed_password}
        password_to_check: Password to verify

    Returns:
        bool: True if password matches, False otherwise
    """
    # Check if email exists in stored logins
    if email not in stored_logins:
        return False

    # Hash the provided password and compare with stored hash
    return stored_logins[email] == hash_password(password_to_check)

def hash_password(password):
    """
    Generates SHA-256 hash of a password

    Args:
        password: Plain text password

    Returns:
        str: Hexadecimal hash string
    """
    return sha256(password.encode()).hexdigest()

def main():
    # Database of stored login credentials (email: hashed_password)
    stored_logins = {
        "example@gmail.com": hash_password("password"),
        "code_in_placer@cip.org": hash_password("Karel"),
        "student@stanford.edu": hash_password("123!456?789")
    }

    # Test cases
    print("Login Attempts:")
    print("example@gmail.com with 'word':",
          login("example@gmail.com", stored_logins, "word"))  # False
    print("example@gmail.com with 'password':",
          login("example@gmail.com", stored_logins, "password"))  # True

    print("\ncode_in_placer@cip.org with 'karel':",
          login("code_in_placer@cip.org", stored_logins, "karel"))  # False
    print("code_in_placer@cip.org with 'Karel':",
          login("code_in_placer@cip.org", stored_logins, "Karel"))  # True

    print("\nstudent@stanford.edu with 'password':",
          login("student@stanford.edu", stored_logins, "password"))  # False
    print("student@stanford.edu with '123!456?789':",
          login("student@stanford.edu", stored_logins, "123!456?789"))  # True

if __name__ == '__main__':
    main()
```
