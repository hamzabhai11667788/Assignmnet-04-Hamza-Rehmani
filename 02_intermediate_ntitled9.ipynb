```python
print("solution")
"""
Planetary Weight Calculator
Calculates weight on different planets based on Earth weight
"""

# Constants for gravity multipliers
MERCURY_GRAVITY = 0.376
VENUS_GRAVITY = 0.889
MARS_GRAVITY = 0.378
JUPITER_GRAVITY = 2.36
SATURN_GRAVITY = 1.081
URANUS_GRAVITY = 0.815
NEPTUNE_GRAVITY = 1.14

def main():
    print("Planetary Weight Calculator")
    print("--------------------------")

    # Get Earth weight input
    earth_weight = float(input("Enter a weight on Earth: "))

    # Get planet input
    planet = input("Enter a planet: ")

    # Determine gravity constant
    if planet == "Mercury":
        gravity_constant = MERCURY_GRAVITY
    elif planet == "Venus":
        gravity_constant = VENUS_GRAVITY
    elif planet == "Mars":
        gravity_constant = MARS_GRAVITY
    elif planet == "Jupiter":
        gravity_constant = JUPITER_GRAVITY
    elif planet == "Saturn":
        gravity_constant = SATURN_GRAVITY
    elif planet == "Uranus":
        gravity_constant = URANUS_GRAVITY
    elif planet == "Neptune":
        gravity_constant = NEPTUNE_GRAVITY
    else:
        print("Invalid planet entered!")
        return

    # Calculate and round weight
    planetary_weight = earth_weight * gravity_constant
    rounded_weight = round(planetary_weight, 2)

    # Format output to remove .0 when not needed
    if rounded_weight == int(rounded_weight):
        print(f"The equivalent weight on {planet}: {int(rounded_weight)}")
    else:
        print(f"The equivalent weight on {planet}: {rounded_weight}")

if __name__ == "__main__":
    main()
```

```python
print("starter_code.")
"""
Planetary Weight Calculator
Prompts the user for a weight on Earth and prints the equivalent weight on Mars or other planets.
"""

def main():
    print("Planetary Weight Calculator")
    print("--------------------------")

    # Get Earth weight input
    try:
        earth_weight = float(input("Enter a weight on Earth: "))
    except ValueError:
        print("Please enter a valid number for weight.")
        return

    # Ask user if they want Mars or other planets
    choice = input("Calculate for Mars only? (yes/no): ").lower()

    if choice == 'yes':
        # Mars-only calculation
        mars_weight = earth_weight * 0.378
        rounded_weight = round(mars_weight, 2)
        print(f"\nThe equivalent weight on Mars: {rounded_weight}")
    else:
        # All planets calculation
        planet = input("Enter a planet (Mercury, Venus, Mars, Jupiter, Saturn, Uranus, Neptune): ")

        # Gravity constants
        gravity_factors = {
            'Mercury': 0.376,
            'Venus': 0.889,
            'Mars': 0.378,
            'Jupiter': 2.36,
            'Saturn': 1.081,
            'Uranus': 0.815,
            'Neptune': 1.14
        }

        if planet in gravity_factors:
            planet_weight = earth_weight * gravity_factors[planet]
            rounded_weight = round(planet_weight, 2)

            # Format output to remove .0 when not needed
            if rounded_weight == int(rounded_weight):
                print(f"\nThe equivalent weight on {planet}: {int(rounded_weight)}")
            else:
                print(f"\nThe equivalent weight on {planet}: {rounded_weight}")
        else:
            print("Invalid planet entered. Please capitalize the first letter and try again.")

if __name__ == "__main__":
    main()
```
